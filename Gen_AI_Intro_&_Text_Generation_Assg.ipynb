**Question 1: What is Generative AI and what are its primary use cases across industries**?

Answer:

Generative Artificial Intelligence (Generative AI) refers to a class of artificial intelligence systems capable of generating new content such as text, images, audio, video, or code by learning patterns from large datasets. **Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation**?

Answer:

Autoencoders and Variational Autoencoders (VAEs) are neural network architectures used for learning compressed representations of data.

**Autoencoders**

An autoencoder consists of two main components:

**Encoder** – compresses input data into a lower-dimensional latent representation.

**Decoder** – reconstructs the original data from the latent representation.

The goal of an autoencoder is to minimize the reconstruction error between the input and output.

However, traditional autoencoders do not model the probability distribution of latent variables, which limits their ability to generate new data.

**Variational Autoencoders (VAEs)**

Variational Autoencoders extend autoencoders by introducing probabilistic latent variables.

Instead of mapping input data to a fixed latent vector, VAEs learn the distribution of latent variables (mean and variance). New samples can be generated by sampling from this distribution.

Key components:

Encoder outputs mean (μ) and variance (σ)

Latent variables are sampled using the reparameterization trick

Decoder reconstructs data from sampled latent variables

**Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical**?

Answer:

Attention mechanisms are an important component in Neural Machine Translation systems that allow models to focus on relevant parts of the input sentence when generating translations.

Traditional encoder-decoder models compress the entire input sentence into a single vector. This often causes information loss, especially for long sentences.

Attention solves this problem by enabling the decoder to dynamically focus on different parts of the input sequence.

**Working of Attention Mechanism**

**Encoder processes input sentence**

Each word is converted into a hidden state.

**Attention scores are calculated**

The decoder compares its current state with encoder states to compute relevance scores.

**Context vector is computed**

Weighted sum of encoder hidden states.

**Decoder generates output**

Uses context vector to generate translated words.

**Why Attention is Critical**

Improves translation quality

Handles long sentences effectively

Allows dynamic alignment between input and output words

Reduces information bottleneck

Attention mechanisms are now fundamental to modern architectures such as transformers used in large language models.

**Question 5: What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling**?

Answer:

The use of generative AI in creative content generation raises several ethical concerns.

**1. Copyright and Intellectual Property**

AI models are trained on large datasets that may include copyrighted material. Generated content may unintentionally resemble existing works, raising legal concerns.

**2. Plagiarism**

AI-generated text could replicate phrases or writing styles from training data, which may lead to plagiarism issues.

**3. Bias and Fair Representation**

If training data contains cultural or social biases, generated stories or poetry may reflect those biases.

**4. Authorship and Ownership**

Determining who owns AI-generated content can be difficult. Questions arise regarding whether the author is the user, the developer, or the AI system.

**5. Misinformation**

Generative AI could produce misleading or false narratives that may influence public perception.

**6. Creative Integrity**

Some critics argue that AI-generated creative works may reduce opportunities for human artists and writers.

To address these concerns, developers must implement ethical guidelines, transparency, and bias mitigation techniques when deploying generative AI systems.

**Question 6: Train a simple VAE for text reconstruction**

Answer:

**Step 1: Dataset**

```python
texts = [
"The sky is blue",
"The sun is bright",
"The grass is green",
"The night is dark",
"The stars are shining"
]
```

**Step 2: Data Preprocessing**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts = [
"The sky is blue",
"The sun is bright",
"The grass is green",
"The night is dark",
"The stars are shining"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
padded = pad_sequences(sequences, padding='post')

print(padded)
```

**Step 3: Build VAE Model**

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

input_dim = padded.shape[1]

inputs = Input(shape=(input_dim,))
encoded = Dense(8, activation='relu')(inputs)
latent = Dense(2)(encoded)

decoded = Dense(8, activation='relu')(latent)
outputs = Dense(input_dim, activation='sigmoid')(decoded)

vae = Model(inputs, outputs)
vae.compile(optimizer='adam', loss='mse')
```

**Step 4: Train Model**

```python
vae.fit(padded, padded, epochs=100)
```
