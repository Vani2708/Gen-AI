**1. What is a Vector Database (VectorDB) and how is it different from traditional databases?**

Ans: A Vector Database (VectorDB) is a specialized database designed to store, index, and search vector embeddings generated by machine learning models. These vectors represent data such as text, images, audio, or video in numerical form, allowing similarity comparisons using mathematical distance metrics like cosine similarity, Euclidean distance, or dot product.

Vector databases are widely used in Artificial Intelligence (AI) applications such as semantic search, recommendation systems, and Retrieval-Augmented Generation (RAG).

Key Features of Vector Databases

Store high-dimensional vector embeddings.

Perform similarity search efficiently.

Use indexing techniques like HNSW (Hierarchical Navigable Small World) or FAISS indexing.

Designed for AI/ML workflows.

**Difference Between Vector Databases and Traditional Databases (Short Answer):**

A **Vector Database** stores data in the form of **vector embeddings**, which are numerical representations generated by AI or machine learning models. It is designed to perform **similarity searches**, meaning it finds results that are semantically or contextually similar to a query. This makes vector databases useful for applications like semantic search, recommendation systems, and AI chatbots.

In contrast, **Traditional Databases** store **structured data in tables with rows and columns** and retrieve information using exact queries or conditions, usually through SQL. They are commonly used for applications such as banking systems, inventory management, and record keeping.

In simple terms, **vector databases focus on similarity-based search for AI applications**, while **traditional databases focus on exact data retrieval for structured information**.
